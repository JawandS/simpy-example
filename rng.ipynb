### Dependencies

```python
import random
import matplotlib.pyplot as plt
```

### Generate rolls

```python
total_rolls = 500
dice_one = [random.randint(1, 6) for _ in range(total_rolls)]
dice_two = [random.randint(1, 6) for _ in range(total_rolls)]
sum_dice = [dice_one[i] + dice_two[i] for i in range(total_rolls)]
```

### Print summary

```python
print(f"Die 1: {sum(dice_one) / total_rolls:.2f}")
print(f"Die 2: {sum(dice_two) / total_rolls:.2f}")
print(f"Sum: {sum(sum_dice) / total_rolls:.2f}")
```

### Graph Results

```python
plt.hist(dice_one, alpha=0.5, bins=range(1, 8), label='Die 2')
plt.title('Histogram of Dice Rolls')
plt.xlabel('Die Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.hist(dice_two, bins=range(1, 8), alpha=0.5, label='Die 2')
plt.title('Histogram of Dice Rolls')
plt.xlabel('Die Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.hist([dice_one[idx] + dice_two[idx] for idx in range(total_rolls)], bins=range(2, 14), alpha=0.5, label='Die 2')
plt.title('Histogram of Dice Rolls')
plt.xlabel('Die Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
